```python
from datetime import datetime

# 将 DateSubmitted 列转换为 datetime，并对时间进行取整到小时
scores_df['DateSubmitted'] = pd.to_datetime(scores_df['DateSubmitted']).dt.tz_localize('utc').dt.tz_convert('Asia/Shanghai')

# 创建一个新的 DataFrame，其中包含每个团队在每个小时的最高分数
# 首先对数据进行分组
grouped_scores = scores_df.groupby(['TeamId', 'TeamName', 'DateSubmitted'])

# 对每个团队和时间段，选取 PublicScore 的最大值
max_scores = grouped_scores['PublicScore'].max().reset_index()

# 重新组织 DataFrame，将时间作为列，团队作为行
pivot_scores = max_scores.pivot(index=['TeamId', 'TeamName'], columns='DateSubmitted', values='PublicScore')
```

```python
pivot_scores.head()
```

```python
def fill_nan_with_max(arr):
    # 初始化最大值为0，因为如果序列开始就是nan，需要用0填充
    max_val = 0
    # 遍历数组中的每个元素
    for i in range(len(arr)):
        if np.isnan(arr[i]):
            # 如果元素是nan，则用当前的最大值替换
            arr[i] = max_val
        else:
            # 更新最大值
            max_val = max(max_val, arr[i])
    return arr

# 示例数组
arr = np.array([np.nan, 3, np.nan, 2, np.nan, 5, np.nan])

# 调用函数并打印结果
fill_nan_with_max(arr)
```

```python
pivot_scores_T = pivot_scores.T
for col in pivot_scores_T:
    pivot_scores_T.loc[:,col] = fill_nan_with_max(pivot_scores_T[col].values)
```

```python
pivot_scores = pivot_scores_T.T
```
